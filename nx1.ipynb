```python
import networkx as nx
import matplotlib.pyplot as plt

# Specify the path to your text file
txt_file = "nx1.txt"

# Create an empty NetworkX graph
G = nx.Graph()

# Read the edge list from the text file and add edges to the graph
with open(txt_file, 'r') as file:
    for line in file:
        source, target = line.strip().split()
        G.add_edge(source, target)

# Calculate node positions using a layout algorithm (e.g., spring_layout)
pos = nx.spring_layout(G)

# Draw the network graph
nx.draw_networkx(G, pos)

# Display the graph
plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# Specify the path to your text file
txt_file = "nxw.txt"

# Create an empty NetworkX directed graph
G = nx.DiGraph()

# Read the edge list from the text file and add directed edges with weights to the graph
with open(txt_file, 'r') as file:
    for line in file:
        # Split the line into parts
        parts = line.strip().split()
        if len(parts) >= 3:
            source, target, weight = parts[:3]
            # Convert the weight to a float (or the appropriate data type)
            weight = float(weight)
            G.add_edge(source, target, weight=weight)  # Add the weight as an attribute
            
weights = [d['weight'] for s, t, d in G.edges(data=True)]
# Calculate node positions using a layout algorithm (e.g., spring_layout)
pos = nx.spring_layout(G)

# Draw the directed network graph with edge weights
ax = plt.subplot(111)
nx.draw_networkx(
    G, pos, width=4, edge_color=weights, edge_cmap=plt.cm.Greys)
ax.margins(0.15, 0.15)

# Display the graph
plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# Specify the path to your text file
txt_file = "nxz.txt"

# Create an empty NetworkX directed graph
G = nx.DiGraph()

# Read the edge list from the text file and add directed edges with weights and colors to the graph
with open(txt_file, 'r') as file:
    for line in file:
        # Split the line into parts
        parts = line.strip().split()
        if len(parts) >= 3:
            source, target, weight = parts[:3]
            # Convert the weight to a float (or the appropriate data type)
            weight = float(weight)
            # Create a dictionary to store edge attributes (weight and color)
            edge_attrs = {'weight': weight}
            if len(parts) >= 4:
                color = parts[3]
                edge_attrs['color'] = color
            G.add_edge(source, target, **edge_attrs)  # Add the edge with attributes

# Extract weights and colors for visualization
weights = [d['weight'] for s, t, d in G.edges(data=True)]
colors = [d.get('color', 'default_color') for s, t, d in G.edges(data=True)]

# Calculate node positions using a layout algorithm (e.g., spring_layout)
pos = nx.spring_layout(G)

# Draw the directed network graph with edge weights and colors
ax = plt.subplot(111)
nx.draw_networkx(
    G, pos, width=4, edge_color=colors, edge_cmap=plt.cm.Greys)
ax.margins(0.15, 0.15)

# Display the graph
plt.show()

```
